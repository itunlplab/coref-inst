```python
import json
```

```python
llm_output = list()

with open('output.jsonl', 'r') as f:
    while True:
        line = f.readline()
        if not line: break
        line = json.loads(line)
        llm_output.append(line)
```

```python
for i in llm_output:
    print(i['text'])
```
