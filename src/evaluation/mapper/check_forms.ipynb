```python
import udapi
```

```python
doc = udapi.Document('dev.conllu')
mapper_doc = udapi.Document('pl_pcc-out.conllu')
```

```python
mentions = []
for ent in doc.coref_entities: 
    mentions.extend([list(map(lambda i: i.form, m.words)) for m in ent.mentions])
```

```python
mentions_mapper = []
for ent in mapper_doc.coref_entities: 
    mentions_mapper.extend([list(map(lambda i: i.form, m.words)) for m in ent.mentions])
```

```python
import copy
for i in copy.deepcopy(mentions_mapper):
    i_form = " ".join(i)

    for j in copy.deepcopy(mentions):
        j_form = " ".join(j)
        if i_form == j_form:
            mentions.remove(i)
            mentions_mapper.remove(j)
            print(i, j)
            break
        #elif i[0] == j[0]:
        #    print(i, j)
```
